### Import Matplotlib 

```python
import matplotlib.pyplot as plt
%matplotlib inline
# additional imports
import numpy as np
```

### Demo
Just a demo line of code to show how plotting is done using matplotlib. Here we are using the pyplot module of matplotlib to plot values.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.grid()
```

### 1. Simple Line Plot
Let's create a simple plot using matplotlib. We will create a list of values for x and corresponding random values for y. And we will plot it using <b>plt.plot(x, y)</b>.

```python
x = [i for i in range(10)] # serializing x axis values
y = [item+np.random.randint(-50, 50)/10 for item in x] # randomizing the y axis values

plt.plot(x, y)
plt.title('Simple Line Plot') # to give the plot a title.
plt.grid() # to print the grid.
plt.show()
```

### 2. Simple Scatter Plot
Let's create a simple scatter plot using matplotlib. We will use the same previously generated values of x and y here. And we will plot it using <b>plt.scatter(x, y)</b>. 

```python
plt.scatter(x, y)
plt.title('Simple Scatter Plot') # to give the plot a title.
plt.grid() # to print the grid.
plt.show()
```

### 3. Visualizing Errors
Let's create a vizualization of error for two list of values using matplotlib. We will use the same previously generated values of x and y here. And we will plot it using <b>plt.errorbar(x, y)</b>. For any scientific measurement, accurate accounting for errors is nearly as important, if not more important, than accurate reporting of the number itself. In visualization of data and results, showing these errors effectively can make a plot
convey much more complete information.

```python
plt.errorbar(x, y, yerr=2, fmt='.k') # xerr, yerr - error margin
plt.title('Error Visualization Plot') # to give the plot a title.
plt.grid() # to print the grid.
plt.show()
```

### 4. Density and Contour Plots
Sometimes it is useful to display three-dimensional data in two dimensions using contours or color-coded regions. There are three Matplotlib functions that can be helpful for this task: plt.contour for contour plots, plt.contourf for filled contour plots, and plt.imshow for showing images.

```python
def get_z(x, y):
    sqr_sum = x**2 + y**2
    return (np.sin(sqr_sum) + np.cos(sqr_sum))

xx = np.linspace(-10, 10, 100)
yy = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(xx, yy)
Z = get_z(X, Y)

plt.contour(X, Y, Z)
plt.title('Plotting Conture of the defined function') # to give the plot a title.
#plt.grid() # to print the grid.
plt.show()
```

```python
plt.contour(X, Y, Z, cmap='RdGy') # using cmap, we can change the plot styles
plt.title('Plotting Conture of the defined function') # to give the plot a title.
#plt.grid() # to print the grid.
plt.show()
```

##### combine contour plots and image plots

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, fontsize=6, inline=True)
plt.imshow(Z, extent=[-2, 2, -2, 2], origin='lower', cmap='RdGy', alpha=0.5)
plt.colorbar()
```

### 5. Multiple Sublots
Let's create a multiple subplots using pyplot. <br>
Here we will create 2X2 subplot with all the plots we created till now. We will use <b>plt.subplot(2, 2, i)</b> where 0 < i <= 4.

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Simple Line Plot') # to give the subplot a title.
plt.grid() # to print the grid.

plt.subplot(2, 2, 2)
plt.scatter(x, y)
plt.title('Simple Scatter Plot') # to give the subplot a title.
plt.grid() # to print the grid.

plt.subplot(2, 2, 3)
plt.errorbar(x, y, yerr=2, fmt='.k') # xerr, yerr - error margin
plt.title('Error Visualization Plot') # to give the subplot a title.
plt.grid() # to print the grid.

plt.subplot(2, 2, 4)
plt.contour(X, Y, Z)
plt.title('Plotting Conture of the defined function') # to give the subplot a title.

plt.suptitle('Combining All The Plots') # common title for all the inputs.

plt.show()
```

### 6. Histogram
Let us plot a histogram using matplotlib.

```python
data = np.random.randn(100) # randomly generated 100 data sample
plt.hist(data) # plotting generated data
plt.title('Plotting Histogram') # to give the subplot a title.
plt.grid() # to print the grid.
plt.show()
```

#### Multiple Histograms

Available histtype values are = ['bar', 'barstacked', 'step',  'stepfilled']<br>
Change it according to your wish and plot.

```python
x1 = np.random.normal(-3, 2, 1000)
x2 = np.random.normal(3, 8, 1000)
x3 = np.random.normal(0, 5, 1000)

args = dict(histtype='stepfilled', alpha=0.8, density=True, bins=20)

plt.hist(x1, label='Type 1', **args)
plt.hist(x2, label='Type 2', **args)
plt.hist(x3, label='Type 3', **args)

plt.title('Plotting multiple histograms') # to give the subplot a title.
plt.grid() # to print the grid.

plt.show()
```

### 7. Visualization With Seaborn
Seaborn provides an API on top of Matplotlib that offers sane choices for plot style and color defaults, defines simple high-level functions for common statistical plot types, and integrates with the functionality provided by Pandas DataFrames.<br>

```python
import pandas as pd
import seaborn as sns

sns.set()
```

Often in statistical data visualization, all we want is to plot histograms. Here we define a dataframe and plot histogram for both of it's columns.

```python
data = np.random.multivariate_normal([0,0], [[7, 2], [2, 2]], size=3000)
data = pd.DataFrame(data, columns=['A', 'B'])

for col in 'AB':
    plt.hist(data[col], density=True, alpha=0.5)
```

Now plotting a smooth estimate of the distribution using a kernel density estimation, which Seaborn does with <b>sns.kdeplot</b>.

```python
for col in 'AB':
    sns.kdeplot(data[col], shade=True)
```

Now, let's combine both the plots in a single figure. 

```python
for col in 'AB':
    plt.hist(data[col], density=True, alpha=0.5)
    sns.kdeplot(data[col], shade=True)
plt.title('Plot using seaborn and matplotlib')
plt.show()
```
